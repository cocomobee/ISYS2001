## Part 4: Enhancement Challenge

Pick ONE feature to add with AI's help:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 A detailed budget summary with advice


```python
# Document your enhancement choice and the prompt you used
enhancement_choice = "I chose to add: input validation"
enhancement_prompt = """
Ok lets add input validation to this code. It will need to warn about entering negative numbers and non-float/integer inputs, and then allow the user to retry to put in different numbers or quit.
"""

print(f"🚀 Enhancement: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
def get_valid_amount(prompt):
    """
    Repeatedly asks the user for a monetary amount until they enter
    a valid, non-negative number. Also allows the user to type 'q' to quit.
    """
    while True:
        user_input = input(prompt)

        # Allow the user to quit at any time
        if user_input.lower() == "q":
            print("Exiting the budget tracker. Goodbye!")
            exit()

        # Check if the input is a valid number
        try:
            amount = float(user_input)
        except ValueError:
            print("Invalid input. Please enter a valid number (or 'q' to quit).")
            continue

        # Check if the number is negative
        if amount < 0:
            print("Amount cannot be negative. Please try again.")
            continue

        # If we get here, the input is valid
        return amount


# Ask the user for their monthly income
monthly_income = get_valid_amount("What is your monthly income? (or 'q' to quit): ")

# Ask for up to 3 expenses
total_expenses = 0

for i in range(1, 4):
    expense = get_valid_amount(f"Enter expense #{i} (or 0 if none, or 'q' to quit): ")
    total_expenses += expense

# Calculate what is left over
remaining_budget = monthly_income - total_expenses

# Print a summary
print(f"\n--- Budget Summary ---")
print(f"Monthly Income: ${monthly_income:.2f}")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")

# Warn the user if the budget goes negative
if remaining_budget < 0:
    print("\nWarning: You have gone over budget!")
    over_budget = total_expenses - monthly_income
    print(f"Over budget by: ${over_budget:.2f}")
else:
    print("\nYou are within your budget.")
```
